```python
X = pd.to_numeric(df.iloc[:, 70], errors='coerce')
y = pd.to_numeric(df.iloc[:, 71], errors='coerce')

plt.scatter(X,y, )
plt.xlabel('Male life expectancy, (2012-14)')
plt.ylabel('Female life expectancy, (2012-14)')
plt.title('Scatter plot of male vs female life expectancy')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```
